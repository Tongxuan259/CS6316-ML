# Model Implementation

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import torchvision.utils as vutils
import matplotlib.pyplot as plt
import numpy as np
import os
from google.colab import drive
%matplotlib inline


# Default MLP with two sigmoid hidden layers and one softmax output layer
# 192 hidden neurons, xavier_uniform weight initialization and zeros bias initialization
class MLP(nn.Module):
    def __init__(self, input_size=784, num_classes=10):
        super(MLP, self).__init__()
        self.input_size = input_size
        self.num_classes = num_classes
        # hidden layer 1
        self.hidden1 = nn.Linear(self.input_size , 128)
        nn.init.xavier_uniform_(self.hidden1.weight)
        nn.init.zeros_(self.hidden1.bias)
        self.sigmoid1 = nn.Sigmoid()
        # hidden layer 2
        self.hidden2 = nn.Linear(128, 64)
        nn.init.xavier_uniform_(self.hidden2.weight)
        nn.init.zeros_(self.hidden2.bias)
        self.sigmoid2 = nn.Sigmoid()
        # ouput layer
        self.output = nn.Linear(64, self.num_classes)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        # flatten input
        x = x.view(-1, self.input_size)
        x = self.sigmoid1(self.hidden1(x))
        x = self.sigmoid2(self.hidden2(x))
        x = self.softmax(self.output(x))
        return x


# MLP with 384 hidden neurons
class MLP1(nn.Module):
    def __init__(self, input_size=784, num_classes=10):
        super(MLP1, self).__init__()
        self.input_size = input_size
        self.num_classes = num_classes
        # hidden layer 1
        self.hidden1 = nn.Linear(self.input_size , 256)
        nn.init.xavier_uniform_(self.hidden1.weight)
        nn.init.zeros_(self.hidden1.bias)
        self.sigmoid1 = nn.Sigmoid()
        # hidden layer 2
        self.hidden2 = nn.Linear(256, 128)
        nn.init.xavier_uniform_(self.hidden2.weight)
        nn.init.zeros_(self.hidden2.bias)
        self.sigmoid2 = nn.Sigmoid()
        # ouput layer
        self.output = nn.Linear(128, self.num_classes)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        # flatten input
        x = x.view(-1, self.input_size)
        x = self.sigmoid1(self.hidden1(x))
        x = self.sigmoid2(self.hidden2(x))
        x = self.softmax(self.output(x))
        return x


# MLP with xavier_normal distribution weight initialization
class MLP2(nn.Module):
    def __init__(self, input_size=784, num_classes=10):
        super(MLP2, self).__init__()
        self.input_size = input_size
        self.num_classes = num_classes
        # hidden layer 1
        self.hidden1 = nn.Linear(self.input_size , 128)
        nn.init.xavier_normal_(self.hidden1.weight)
        nn.init.zeros_(self.hidden1.bias)
        self.sigmoid1 = nn.Sigmoid()
        # hidden layer 2
        self.hidden2 = nn.Linear(128, 64)
        nn.init.xavier_normal_(self.hidden2.weight)
        nn.init.zeros_(self.hidden2.bias)
        self.sigmoid2 = nn.Sigmoid()
        # ouput layer
        self.output = nn.Linear(64, self.num_classes)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        # flatten input
        x = x.view(-1, self.input_size)
        x = self.sigmoid1(self.hidden1(x))
        x = self.sigmoid2(self.hidden2(x))
        x = self.softmax(self.output(x))
        return x


# MLP with kaiming_uniform distribution weight initialization
class MLP3(nn.Module):
    def __init__(self, input_size=784, num_classes=10):
        super(MLP3, self).__init__()
        self.input_size = input_size
        self.num_classes = num_classes
        # hidden layer 1
        self.hidden1 = nn.Linear(self.input_size , 128)
        nn.init.kaiming_uniform_(self.hidden1.weight)
        nn.init.zeros_(self.hidden1.bias)
        self.sigmoid1 = nn.Sigmoid()
        # hidden layer 2
        self.hidden2 = nn.Linear(128, 64)
        nn.init.kaiming_uniform_(self.hidden2.weight)
        nn.init.zeros_(self.hidden2.bias)
        self.sigmoid2 = nn.Sigmoid()
        # ouput layer
        self.output = nn.Linear(64, self.num_classes)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        # flatten input
        x = x.view(-1, self.input_size)
        x = self.sigmoid1(self.hidden1(x))
        x = self.sigmoid2(self.hidden2(x))
        x = self.softmax(self.output(x))
        return x


# MLP with kaiming_normal distribution weight initialization
class MLP4(nn.Module):
    def __init__(self, input_size=784, num_classes=10):
        super(MLP4, self).__init__()
        self.input_size = input_size
        self.num_classes = num_classes
        # hidden layer 1
        self.hidden1 = nn.Linear(self.input_size , 128)
        nn.init.kaiming_normal_(self.hidden1.weight)
        nn.init.zeros_(self.hidden1.bias)
        self.sigmoid1 = nn.Sigmoid()
        # hidden layer 2
        self.hidden2 = nn.Linear(128, 64)
        nn.init.kaiming_normal_(self.hidden2.weight)
        nn.init.zeros_(self.hidden2.bias)
        self.sigmoid2 = nn.Sigmoid()
        # ouput layer
        self.output = nn.Linear(64, self.num_classes)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        # flatten input
        x = x.view(-1, self.input_size)
        x = self.sigmoid1(self.hidden1(x))
        x = self.sigmoid2(self.hidden2(x))
        x = self.softmax(self.output(x))
        return x


# MLP with uniform distribution bias initialization
class MLP5(nn.Module):
    def __init__(self, input_size=784, num_classes=10):
        super(MLP5, self).__init__()
        self.input_size = input_size
        self.num_classes = num_classes
        # hidden layer 1
        self.hidden1 = nn.Linear(self.input_size , 128)
        nn.init.xavier_uniform_(self.hidden1.weight)
        nn.init.uniform_(self.hidden1.bias)
        self.sigmoid1 = nn.Sigmoid()
        # hidden layer 2
        self.hidden2 = nn.Linear(128, 64)
        nn.init.xavier_uniform_(self.hidden2.weight)
        nn.init.uniform_(self.hidden2.bias)
        self.sigmoid2 = nn.Sigmoid()
        # ouput layer
        self.output = nn.Linear(64, self.num_classes)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        # flatten input
        x = x.view(-1, self.input_size)
        x = self.sigmoid1(self.hidden1(x))
        x = self.sigmoid2(self.hidden2(x))
        x = self.softmax(self.output(x))
        return x


# MLP with normal distribution bias initialization
class MLP6(nn.Module):
    def __init__(self, input_size=784, num_classes=10):
        super(MLP6, self).__init__()
        self.input_size = input_size
        self.num_classes = num_classes
        # hidden layer 1
        self.hidden1 = nn.Linear(self.input_size , 128)
        nn.init.xavier_uniform_(self.hidden1.weight)
        nn.init.normal_(self.hidden1.bias)
        self.sigmoid1 = nn.Sigmoid()
        # hidden layer 2
        self.hidden2 = nn.Linear(128, 64)
        nn.init.xavier_uniform_(self.hidden2.weight)
        nn.init.normal_(self.hidden2.bias)
        self.sigmoid2 = nn.Sigmoid()
        # ouput layer
        self.output = nn.Linear(64, self.num_classes)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        # flatten input
        x = x.view(-1, self.input_size)
        x = self.sigmoid1(self.hidden1(x))
        x = self.sigmoid2(self.hidden2(x))
        x = self.softmax(self.output(x))
        return x


# Convolutional neural network
class CNN(nn.Module):
    def __init__(self, num_classes=10):
        super(CNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 64, kernel_size=3, stride=2, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 192, kernel_size=5, padding=2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(192, 384, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(384, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.avgpool = nn.AdaptiveAvgPool2d((6, 6))
        self.classifier = nn.Sequential(
            nn.Dropout(),
            nn.Linear(256 * 6 * 6, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(),
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            nn.Linear(4096, num_classes),
        )

    def forward(self, x):
        x = self.features(x)
        x = self.avgpool(x)
        x = x.view(x.size(0), 256 * 6 * 6)
        x = self.classifier(x)
        return x
```

```python
# training function
def train_model(train_set, model, optimizer, batch_size, epochs):
    loss_function = nn.CrossEntropyLoss()
    train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
    print('Training {}...'.format(model.__class__.__name__))
    train_loss = []
    for epoch in range(epochs):
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = loss_function(outputs, labels)
            loss.backward()
            optimizer.step()
        train_loss.append(loss.item())
        print(f'Epoch [{epoch+1}/{epochs}], Cross Entropy Loss: {loss.item():.4f}')
    print('{} training complete!'.format(model.__class__.__name__))
    print('-'*100)
    return train_loss

# testing function
def test_model(test_set, model):
    #summary(model, (1, 28, 28))
    test_loader = DataLoader(test_set, batch_size=128, shuffle=True)
    # calculate accuracy
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predictions = torch.max(outputs.data, 1)
            all_preds.extend(predictions.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    accuracy = accuracy_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds, average='macro')
    recall = recall_score(all_labels, all_preds, average='macro')
    f1 = f1_score(all_labels, all_preds, average='macro')
    return accuracy, precision, recall, f1

# load MNIST dataset
train_set = MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())
test_set = MNIST(root='./data', train=False, download=True, transform=transforms.ToTensor())
# check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print('Device: {}'.format(device))
epochs_list = [i for i in range(5, 51, 5)]
drive.mount('/content/drive')
```

## 3.2 Training data augmentation

```python
def plot_images(batch, rows, cols, title = "", ax= None):
    if ax is not None:
        ax.axis("off")
        ax.set_title(title)
        ax.imshow(np.transpose(
            vutils.make_grid(batch[:(rows * cols)], nrow = rows, normalize = True).cpu(),
            (1, 2, 0)))
    else:
        plt.figure(figsize = (rows, cols))
        plt.axis("off")
        plt.title(title)
        plt.imshow(np.transpose(
            vutils.make_grid(batch[:(rows * cols)], nrow = rows, normalize = True).cpu(),
            (1, 2, 0)))

# augmented training data
augment_transform = transforms.Compose([
    transforms.RandomAffine(degrees=(-7, 7),
                            translate=(0.05, 0.05),
                            scale=(0.9, 1.1),
                            shear=(-10, 10)),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))  # Normalizing with MNIST mean and std dev
    ])
augment_train_set = MNIST(root='./data', train=True, download=True, transform=augment_transform)

batch_size = 128
train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=False)
first_batch = next(iter(train_loader))
plot_images(first_batch[0], 16, 8, "Some Original Training Data")

train_loader = DataLoader(augment_train_set, batch_size=batch_size, shuffle=False)
first_batch = next(iter(train_loader))
plot_images(first_batch[0], 16, 8, "Training Data after Augmentation")
```

```python
# training
for epochs in epochs_list:
    mlp = MLP().to(device)
    optimizer = optim.Adam(mlp.parameters(), lr=0.0001)
    train_loss = train_model(augment_train_set, mlp, optimizer, 128, epochs)
    #torch.save(mlp.state_dict(), '/content/drive/My Drive/ML/project/MLP_epochs_{}_augmented.pth'.format(epochs))
    torch.save(mlp.state_dict(), 'MLP_epochs_{}_augmented.pth'.format(epochs))
```
